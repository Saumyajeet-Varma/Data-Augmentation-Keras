# Import Libraries

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

# Loading the Image

```python
img = image.load_img('../Images/Original_Image/cat.jpeg')
```

```python
img
```

# Data Augmentation

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    vertical_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2
)
```

```python
img = image.img_to_array(img)
```

```python
input_batch = img.reshape(1, 464, 464, 3)
```

```python
img_count = 0
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='../Images/Augmented_Images/'):
    img_count += 1
    if img_count == 10:
        break
```
